```python
import py_midicsv

midi_row_list = []
# Convert MIDI to text (CSV format)
midi_file = "../DoReMi_v1/MIDI/accidental tucking - Full score - accidental tucking.mid"
i=0
with open(midi_file, "rb") as midi_in:
    print('hi')
    for row in py_midicsv.midi_to_csv(midi_in):
        cleaned_row = row.strip()
        # print('cleaned row', cleaned_row)
        midi_row_list.append(cleaned_row)
        i+=1
        # print('midi_row_list', midi_row_list)
print(i)
```

```python
for row in midi_row_list:
    print(row)
```
